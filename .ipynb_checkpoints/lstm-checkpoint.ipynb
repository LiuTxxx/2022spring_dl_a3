```python
from Part1.train import *
import argparse
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
config = get_config()
```

```python
T = range(5, 20, 2)
acc_list = []
for t in T:
    config.input_length = t
    acc = train(config, False)
    acc_list.append(acc)
```

```python
plt.title("lstm accuracy")
plt.xlabel("Input length T")
plt.ylabel("Accuracy")
plt.plot(T, acc_list, '.-')
```

```python
config.input_length = 17
LR = [0.001, 0.004, 0.007, 0.01]
acc_list = []
for lr in LR:
    config.lr = lr
    acc = train(config, False)
    print(f'and with lr = {lr}.')
    acc_list.append(acc)
```

```python
plt.title("lstm accuracy")
plt.xlabel("Learning Rate")
plt.ylabel("Accuracy")
plt.plot(LR, acc_list, '.-')
```
